```r
counts <- rbind(table(data$Fatal, data$Rank), total = unlist(table(data$Rank)))
props <- rbind(no_prop = counts[1,] / counts[3,], yes_prop = counts[2,] / counts[3,])
rank_counts <- as.data.frame(cbind(t(counts), t(props)))
rank_counts$rank <- row.names(rank_counts)
rank_counts$overall_no_prop <- rank_counts$No/sum(rank_counts$No)
rank_counts$overall_yes_prop <- rank_counts$Yes/sum(rank_counts$Yes)
rank_counts$overall_no_prop_norm <- rank_counts$No/sum(rank_counts$No)*(rank_counts$total/sum(rank_counts$total))
rank_counts$overall_yes_prop_norm <- rank_counts$Yes/sum(rank_counts$Yes)*(rank_counts$total/sum(rank_counts$total))
rank_counts[order(rank_counts$yes_prop, decreasing = TRUE),]
rank_counts[order(rank_counts$Yes, decreasing = TRUE),]
```

```r
x_labels <- factor(rank_counts[order(rank_counts$yes_prop, decreasing = TRUE),]$rank, ordered = TRUE, levels = rank_counts[order(rank_counts$yes_prop, decreasing = FALSE),]$rank)
ggplot(rank_counts, aes(x = x_labels, y = rank_counts[order(rank_counts$yes_prop, decreasing = TRUE),]$yes_prop,
                         fill = rank_counts[order(rank_counts$yes_prop, decreasing = TRUE),]$overall_yes_prop_norm)) +
    geom_bar(stat = "identity") + scale_y_continuous(labels = scales::percent) + 
  theme_minimal() + ggtitle("Proportion of Fatal Officer Involved Shootings by Rank") +
    xlab("Rank") + ylab("Proportion of Fatal") + coord_flip() +
  scale_fill_gradient("Overall Proportion of \nFatalities Normalized*", low="light blue", high="dark blue")

x_labels <- factor(rank_counts[order(rank_counts$yes_prop, decreasing = TRUE),]$rank, ordered = TRUE, levels = rank_counts[order(rank_counts$yes_prop, decreasing = FALSE),]$rank)
ggplot(rank_counts, aes(x = x_labels, y = rank_counts[order(rank_counts$yes_prop, decreasing = TRUE),]$yes_prop,
                         fill = rank_counts[order(rank_counts$yes_prop, decreasing = TRUE),]$overall_yes_prop_norm)) +
    geom_bar(stat = "identity") + scale_y_continuous(labels = scales::percent) + 
  theme_minimal() +
    theme(axis.text.x = element_text(angle = 90, hjust = 1)) + ggtitle("Fatal Officer Involved Shootings by Rank") +
    xlab("Rank") + ylab("Proportion of Fatal") +
  scale_fill_gradient("Overall Proportion of \nFatalities Normalized*", low="light blue", high="dark blue")
```

```r
ggplot(data = data, aes(fill = Rank, Fatal)) + geom_bar(position="stack") + theme_minimal() +
  ylab("Count") + ggtitle("Count of Fatal Officer Involved Shootings by Rank") 
ggplot(data = data, aes(fill = Rank, Fatal)) + geom_bar(position="fill") + theme_minimal() +
  ylab("Proportion") + ggtitle("Proportion of Fatal Officer Involved Shootings by Rank")
```

```r
ggplot(data = data, aes(fill = Fatal, x = forcats::fct_infreq(Rank))) + 
  geom_bar(position = "dodge") + theme_minimal() + coord_flip() +
  xlab("Rank") + ylab("Count") + ggtitle("Count of Fatal Officer Involved Shootings by Rank")

ggplot(data = data, aes(fill = Fatal, x = factor(data$Rank, levels = levels(x_labels)))) + 
  geom_bar(position = "stack") + theme_minimal() + coord_flip() +
  xlab("Rank") + ylab("Count of Reports") + ggtitle("Reports of Fatal Officer Involved Shootings by Rank")

ggplot(data = data, aes(fill = Fatal, x = forcats::fct_infreq(Rank))) + 
  geom_bar(position = "fill") + theme_minimal() + coord_flip() +
  xlab("Rank") + ylab("Proportion") + ggtitle("Proportion of Fatal Officer Involved Shootings by Rank")
```
