```python
# remove duplicated reported incidents
clean_df = data.drop_duplicates('Incident Number')
clean_df.head(10)
```

```python
minLon = min(clean_df['Longitude'])
maxLon = max(clean_df['Longitude'])
minLat = min(clean_df['Latitude'])
maxLat = max(clean_df['Latitude'])
```

```python
Lon = np.arange(minLon, maxLon, (maxLon-minLon)/100) 
Lat = np.arange(minLat, maxLat, (maxLat-minLat)/100) 
```

```python
Crime_counts = np.zeros((100,100)) 
```

```python
for a in range(len(clean_df)):
    for b1 in range(100):
        if Lat[b1] - (maxLat-minLat)/200 <= clean_df['Latitude'].values[a] < Lat[b1] + (maxLat-minLat)/200:
            for b2 in range(100):
                if Lon[b2] - (maxLon-minLon)/200 <= clean_df['Longitude'].values[a] < Lon[b2] + (maxLon-minLon)/200:
                    Crime_counts[b1,b2] += 1
```
